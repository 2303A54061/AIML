```python
people = ['Ann','Bradon','Chenn','David','Emily','Farook',
          'Gagan','Hamish','Imran','Julio','Katherine','Lily']
age = [21,12,32,45,17,18,28,52,5,40,48,15]
weight = [55,35,77,68,70,60,72,69,18,645,82,48]
height = [100,135,170,165,173,168,175,159,105,171,155,158]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("Height in meters")
plt.xlabel("Weight of patients in kgs")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs", fontsize=16, fontstyle='italic')

#main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6, color='orange', edgecolor='k', alpha=0.6)
#edgecolor='k': sets the edge color of the bars to black
#alpha=0.6: sets the transparancy of the bars, with 1 being fully opaque and 0 fully transparent.

plt.xlabel("People", fontsize=15)
plt.ylabel("Weight (in kgs)", fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))

plt.hist(weight,color='red', edgecolor='k', alpha=0.75, bins=5)
 #bins=5 : the number of bins(intervals) to divide the data range into.

plt.title("Histogram of patient weight", fontsize=18)
plt.xlabel("Weight in kgs", fontsize=15)
plt.show()
```
